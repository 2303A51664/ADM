```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/diabetes.csv')
plt.hist(df['Age'],bins=40,color='skyblue',edgecolor='black')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
summary=df.describe()
print(summary)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Glucose'], y=df['Insulin'], color='purple')
plt.title('Glucose vs. Insulin', fontsize=14)
plt.xlabel('Glucose', fontsize=12)
plt.ylabel('Insulin', fontsize=12)
plt.show()
```
